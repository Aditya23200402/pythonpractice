Laptop Sales

```python
import matplotlib.pyplot as plt

def percent(sold,company):
    li=[]
    sum1=sum(sold)
    for i in range(len(sold)):
        num = (sold[i]*100)/sum1
        print(company[i], round(num,1))
        li.append(num)
    s = str(li)
    return s


company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]

plt.pie(sold, labels=company ,explode=[0.1,0.2,0.1,0.2,0.1,0], autopct='%.1f', counterclock=False, startangle=90)
plt.show()
st1 = percent(sold, company)
print(st1)
```

Plot Job Year

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('amazon_jobs_dataset.csv')
df['Posting_date'] = df['Posting_date'].str[-4:]

data = {}
for i in df.Posting_date:
    if i in data: 
       data[i] += 1
    else:
        data[i] = 1

for key in data:
    print(key, data[key])
year = data.keys()

print(year)
no_ofjobs = data.values()

plt.plot(year,no_ofjobs)
plt.show()
```

Plot Jobs Months

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt    

df = pd.read_csv('amazon_jobs_dataset.csv')
df['Posting_date'] = df['Posting_date'].str.split(' ').str[0]
df = df.groupby(by=['Posting_date']).size()
Month =[]
no_of_jobs=[]
for i in df.index:
    Month.append(i)
    no_of_jobs.append(df.loc[i])
    print(i,df.loc[i])

plt.bar(Month, no_of_jobs)
plt.xticks(rotation=45)
plt.show()
```

Plot City_Job

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

df = pd.read_csv('amazon_jobs_dataset.csv')
df.location=df['IN' == df['location'].str.split(', ').str[0]].location.str.split(', ').str[2]
df.dropna()
group = df.groupby(by='location')

dic={}
for i,j in group:
    dic[i]=df.location.str.count(i).sum()
df = group.size()
df = df.sort_values(ascending=False)
City = []
No_of_job=[]
for i in df.index:
    City.append(i)
    No_of_job.append(df.loc[i])
    print(i,format(df.loc[i]*100/sum(dic.values()),'.2f'))
plt.pie(No_of_job,labels=City,autopct='%.2f%%')
plt.show()
    
```

Plot Job Java

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import csv

with open('amazon_jobs_dataset.csv', encoding='utf8') as file_obj:
    file_data=csv.DictReader(file_obj, skipinitialspace=True)
    year=[]

    for row in file_data:
        if 'java' in row['BASIC QUALIFICATIONS'] or 'Java' in row['BASIC QUALIFICATIONS']:
            year.append(row['Posting_date'].split()[2])
    np_year = np.array(year, dtype=int)
    dic={}
    for i in np_year:
        if i in dic:
            dic[i] += 1
        else:
            dic[i] = 1
    
    x = list(dic.keys())
    y = list(dic.values())

    np_x= np.array(x)
    np_y= np.array(y)

    np_x = np_x[::-1]
    np_y= np_y[::-1]

    plt.scatter(np_x, np_y)
    plt.show()
    for i in range(len(np_x)):
        print(np_x[i], np_y[i])
```
